````python
import json
import re

def remove_links(text):

    # Regex pattern to match URLs
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    # Remove URLs from the text
    cleaned_text = url_pattern.sub('', text)
    return cleaned_text.strip()

# Load the rawData.json file
input_file = "rawData.json"
output_file = "NoLinks.json"

# Read the JSON data
with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)  # Load JSON data

# Clean each entry in the JSON data
cleaned_data = []
for entry in data:
    if "text" in entry:
        cleaned_entry = {"text": remove_links(entry["text"])}
    else:
        cleaned_entry = entry  # Keep other fields unchanged
    cleaned_data.append(cleaned_entry)

# Save the cleaned data to a new JSON file
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(cleaned_data, f, ensure_ascii=False, indent=4)

print(f"Cleaned data saved to {output_file}")
````

````python
import json
from transformers import AutoTokenizer

# Load the AraBERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")

def segment_text_into_sentences(text):

    tokens = tokenizer.tokenize(text)
    sentences = []
    current_tokens = []

    # Define boundary tokens
    boundaries = {"،", ".", ";", ":", "?", "!", "؟", "او", "أو", "ثم", "لكن", "لذا", "أمّا",
                  "بينما", "بسبب", "لذلك", "عند", "حسب", "مثل", "ولكن", "في", "على", "قد", "كما"}
    for token in tokens:
        # Skip unknown tokens
        if token == "[UNK]":
            continue

        current_tokens.append(token)

        # If the token is a boundary, convert current tokens to a sentence
        if token in boundaries:
            sentence = tokenizer.convert_tokens_to_string(current_tokens).strip()
            if sentence:
                sentences.append(sentence)
            current_tokens = []

    # Append any remaining tokens as the last sentence
    if current_tokens:
        sentence = tokenizer.convert_tokens_to_string(current_tokens).strip()
        if sentence:
            sentences.append(sentence)

    return sentences

# File paths for your final cleaned data and the segmented output
input_file = "FinalCleanedArabic.json"
output_file = "SegmentedSentences.json"

# Load your final cleaned JSON data (each item is expected to have a "text" field)
with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# Process each text entry to segment it into sentences
segmented_sentences = []
for item in data:
    if "text" in item:
        sentences = segment_text_into_sentences(item["text"])
        segmented_sentences.extend(sentences)
    else:
        print("Warning: missing 'text' field in an item.")

# Save the segmented sentences into a new JSON file in a list format
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(segmented_sentences, f, ensure_ascii=False, indent=4)

print(f"Segmented sentences saved to {output_file}")
````

````python
import json
from camel_tools.tokenizers.word import simple_word_tokenize
from camel_tools.utils.charsets import AR_DIAC_CHARSET

# File paths
input_file = "SegmentedSentences.json"
output_file = "TokenizedWords.json"

# Define words that should stay as a single token
no_split_words = {
    ("و", "راثة"), ("ب", "عض"), ("و", "فاة"), ("و", "حده"), ("و", "قت"),
    ("و", "زن"), ("ب", "دنية"), ("ب", "ين"), ("و", "حدها"), ("و", "ضع"),
    ("ب", "عضها"), ("ب", "عد"), ("ب", "ؤرية"), ("ب", "ينما"), ("ب", "قية"),
    ("و", "ظائف"), ("ب", "داية"), ("ب", "الغين"), ("ب", "يئة"),
    ("و", "جود"), ("ب", "رامج"), ("و", "صول"), ("و", "احدة"), ("و", "اضح")
}

# Define a dictionary for corrupted token corrections
corrupted_token_corrections = {
    "الجندرومع": ("الجندر", "ومع"), "الأزماتهناك": ("الأزمات", "هناك"),
    "تترافققد": ("تترافق", "قد"), "الطبيةوبالتالي": ("الطبية", "وبالتالي"),
    "خطورتهابالنسبة": ("خطورتها", "بالنسبة"), "الطفولةيشتكي": ("الطفولة", "يشتكي"),
    "الجنسيعندما": ("الجنسي", "عندما"), "الولادةتتضمن": ("الولادة", "تتضمن"),
    "الجندرقد": ("الجندر", "قد"), "التوافقالرضا": ("التوافق", "الرضا"),
    "الجندريولكن": ("الجندري", "ولكن"), "للجندرتكون": ("للجندر", "تكون"),
    "يحددونهولا": ("يحددونه", "ولا"), "مرضيةوغالبا": ("مرضية", "وغالبا"),
    "معينوقد": ("معين", "وقد"), "الشدةومن": ("الشدة", "ومن"),
    "سهولةومن": ("سهولة", "ومن"), "الأمثلويشمل": ("الأمثل", "ويشمل"),
    "للقلقويمكن": ("للقلق", "ويمكن"), "العيادةوفي": ("العيادة", "وفي"),
    "النوعيةفعلى": ("النوعية", "فعلى"), "الطيرانيمكن": ("الطيران", "يمكن"),
    "الأسئلةولكنهم": ("الأسئلة", "ولكنهم"), "الخصيتينبالنسبة": ("الخصيتين", "بالنسبة"),
    "حياتهمبمجرد": ("حياتهم", "بمجرد"), "الجندرويواجه": ("الجندري", "ويواجه"),
    "الماضيإذا": ("الماضي", "إذا"), "الصلةبالنسبة": ("الصلة", "بالنسبة"),
    "جندريعلى": ("جندري", "على"), "الاجتماعيوقد": ("الاجتماعي", "وقد"),
    "للجندرولكن": ("للجندر", "ولكن"), "الاستسقاءويشتكي": ("الاستسقاء", "ويشتكي"),
    "أوزانهموقد": ("أوزانهم", "وقد"), "الشديدكما": ("الشديد", "كما"),
    "المساعدةيتسم": ("المساعدة", "يتسم"), "الاضطرابوقد": ("الاضطراب", "وقد"),
    "الصحيةورغم": ("الصحية", "ورغم"), "للشخصويشتمل": ("للشخص", "ويشتمل"),
    "العائليوعادة": ("العائلي", "وعادة"), "الاكتئابولمعرفة": ("الاكتئاب", "ولمعرفة"),
    "يمتلكونهاتتراكم": ("يمتلكونها", "تتراكم"), "المحتملةلا": ("المحتملة", "لا"),
    "الأعراضوقد": ("الأعراض", "وقد"), "خطيرويميل": ("خطير", "ويميل"),
    "جوانبهاوقد": ("جوانبها", "وقد"), "رغبتهموقد": ("رغبتهم", "وقد"),
    "عليهوعندما": ("عليه", "وعندما"), "الجدويكون": ("الجد", "ويكون"),
    "الوقتوقد": ("الوقت", "وقد"), "الرعايةقد": ("الرعاية", "قد"),
    "الطبيبيمكن": ("الطبيب", "يمكن"), "لهاولكن": ("لها", "ولكن"),
    "آخروعادة": ("آخر", "وعادة"), "مونخهاوزن": ("مونخها", "وزن"),
    "جيدةقد": ("جيدة", "قد"), "المرضويسعى": ("المرض", "ويسعى"),
    "للطفلعادة": ("للطفل", "عادة"), "مؤكديتعرض": ("مؤكد", "يتعرض"),
    "السلوكعلى": ("السلوك", "على"), "ميديكير": ("ميديكير",),
    "مينيسوتا": ("مينيسوتا",), "فلديك": ("فلديك",), "فالأطفال": ("فالأطفال",),
    "الأطفالتضمنت": ("الأطفال", "تضمنت"),
    "مناسبةينطوي": ("مناسبة", "ينطوي")
}

def remove_diacritics(text):
    return ''.join(c for c in text if c not in AR_DIAC_CHARSET)

def custom_tokenize(sentence):
    sentence = remove_diacritics(sentence)
    tokens = simple_word_tokenize(sentence)
    final_tokens = []
    i = 0
    while i < len(tokens):
        token = tokens[i]

        # Handle predefined no-split words
        if i < len(tokens) - 1 and (token, tokens[i + 1]) in no_split_words:
            final_tokens.append(token + tokens[i + 1])
            i += 2
            continue

        # Handle corrupted token corrections
        if token in corrupted_token_corrections:
            correction = corrupted_token_corrections[token]
            final_tokens.extend(correction)
            i += 1
            continue

        # Handle prefix cases for "و" and "ب"
        if token.startswith("و") and len(token) > 1 and ("و", token[1:]) not in no_split_words:
            final_tokens.append("و")
            final_tokens.append(token[1:])
        elif token.startswith("ب") and len(token) > 1 and ("ب", token[1:]) not in no_split_words:
            final_tokens.append("ب")
            final_tokens.append(token[1:])
        else:
            final_tokens.append(token)

        i += 1
    return final_tokens

# Read input JSON file
with open(input_file, "r", encoding="utf-8") as f:
    segmented_sentences = json.load(f)

# Process sentences and apply custom tokenization
tokenized_sentences = [custom_tokenize(sentence) for sentence in segmented_sentences]

# Write output JSON file
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(tokenized_sentences, f, ensure_ascii=False, indent=4)

print(f"Tokenized sentences saved to {output_file}")
````

````python
import json

def correct_tokenization(annotations):
    # Dictionary for splitting combined words
    split_corrections = {
        "الألمقد": ["الألم", "قد"],
        "الهزيمةوقد": ["الهزيمة", "وقد"],
        "التعافييتعافى": ["التعافي", "يتعافى"],
        "شائعينيخاف": ["شائعين", "يخاف"],
        "أنفسهمويكون": ["أنفسهم", "ويكون"],
        "طبيعيخلال": ["طبيعي", "خلال"],
        "الجارحةتؤدي": ["الجارحة", "تؤدي"],
        "المجتمعتتوفر": ["المجتمع", "تتوفر"],
        "البلوغقد": ["البلوغ", "قد"],
        "معاييرولكن": ["معايير", "ولكن"],
        "للمزاجوقد": ["للمزاج", "وقد"],
        "المريضينبغي": ["المريضين", "يجب"],
        "الانتحاريتنطوي": ["الانتحار", "يتنطوي"],
        "أكثريراقب": ["أكثر", "يراقب"],
        "توبيخاتهيميل": ["توبيخاته", "يميل"],
        "المشاكلوتشتمل": ["المشاكل", "وتشتمل"],
        "عليهاولكن": ["عليها", "ولكن"],
        "توريتولكن": ["توريت", "ولكن"],
        "عادةولكن": ["عادة", "ولكن"],
        "قلقوتشتمل": ["قلق", "وتشتمل"],
        "الرئيسقد": ["الرئيس", "قد"],
        "لديهولكن": ["لديه", "ولكن"],
        "الطيرانيمكن": ["الطيران", "يمكن"],
        "لايكونونقادرينعلى": ["لا", "يكونون", "قادرين", "على"]
    }

    # Dictionary for combining split tokens
    combine_corrections = {
        ("ب", "صرية"): "بصرية",
        ("و", "سيلة"): "وسيلة",
        ("و", "اضحة"): "واضحة",
        ("و", "همية"): "وهمية",
        ("و", "راثي"): "وراثي",
        ("و", "عي"): "وعي",
        ("و", "هم"): "وهم",
        ("و", "اقعه"): "واقعه",
        ("و", "راء"): "وراء",
        ("و", "سيط"): "وسيط",
        ("ب", "ناء"): "بناء",
        ("و", "رم"): "ورم",
        ("ب", "صبصة"): "بصبصة",
        ("و", "اقعيين"): "واقعيين",
        ("ب", "زل"): "بزل",
        ("ب", "طيئين"): "بطيئين",
        ("و", "جه"): "وجه",
        ("ب", "اهت"): "باهت",
        ("و", "سط"): "وسط",
        ("و", "باء"): "وباء",
        ("ب", "ينها"): "بينها",
        ("ب", "يئتين"): "بيئتين",
        ("ب", "ناء"): "بناء"
    }

    corrected_annotations = []
    i = 0
    while i < len(annotations):
        word = annotations[i]["word"].strip()
        tag = annotations[i]["tag"]

        # Check if the current token and the next token should be combined
        if i < len(annotations) - 1:
            next_word = annotations[i + 1]["word"].strip()
            if (word, next_word) in combine_corrections:
                combined_word = combine_corrections[(word, next_word)]
                corrected_annotations.append({"word": combined_word, "tag": tag})
                i += 2  # Skip the next token since it's combined
                continue

        # Check if the current word needs to be split
        if word in split_corrections:
            for w in split_corrections[word]:
                corrected_annotations.append({"word": w, "tag": tag})
        else:
            corrected_annotations.append({"word": word, "tag": tag})

        i += 1

    return corrected_annotations

def process_json_file(input_file, output_file):
    with open(input_file, "r", encoding="utf-8") as file:
        data = json.load(file)

    for entry in data:
        if "annotations" in entry:
            entry["annotations"] = correct_tokenization(entry["annotations"])

    with open(output_file, "w", encoding="utf-8") as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

    print(f"Processed file saved as {output_file}")

# Example usage
input_file = "annotated-data'.json"
output_file = "annotated-data''.json"
process_json_file(input_file, output_file)
````
